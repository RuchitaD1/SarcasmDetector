```
def punctuation(string): 
  
    # punctuation marks 
    punctuations = '''!()-[]{};:'"\,<>./?$%^&*_~'''
  
    # traverse the given string and if any punctuation 
    # marks occur replace it with null 
    for x in string.lower(): 
        if x in punctuations: 
            string = string.replace(x, "") 
    return string
```

```

def build_subj_dicionary(lines):
    subj_dict = dict()
    for line in lines:
        splits = line.split(' ')
        if len(splits) == 6:
            word = splits[2][6:]        # the word analyzed
            word_type = splits[0][5:]   # weak or strong subjective
            pos = splits[3][5:]         # part of speech: noun, verb, adj, adv or anypos
            polarity = splits[5][14:]   # its polarity: can be positive, negative or neutral
            new_dict_entry = polarity
            if word not in subj_dict.keys():
                
                subj_dict[word] = new_dict_entry
    return subj_dict


def get_subj_lexicon():
    lexicon = load_file("mpqa.tff")
    subj_dict = build_subj_dicionary(lexicon)
    return subj_dict
```

```
#V+V, V+ADV, ADV+V, to+V,V+NOUN,  V+PRO,  V+ADJ
patterns = [('V','V'),('V','R'),('R','V'),('V','N'),('V','O'),('V','A')]
```

```

def checkgram2(gram):
  if g[0][0][0] == "to":
    if g[1][0][1] == 'V':

      return True
  else:
    if (g[0][0][1],g[1][0][1]) in patterns:
      return True
  return False
```
