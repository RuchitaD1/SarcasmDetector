```

def build_subj_dicionary(lines):
    subj_dict = dict()
    for line in lines:
        splits = line.split(' ')
        if len(splits) == 6:
            word = splits[2][6:]        # the word analyzed
            word_type = splits[0][5:]   # weak or strong subjective
            pos = splits[3][5:]         # part of speech: noun, verb, adj, adv or anypos
            polarity = splits[5][14:]   # its polarity: can be positive, negative or neutral
            new_dict_entry = polarity
            if word not in subj_dict.keys():
                
                subj_dict[word] = new_dict_entry
    return subj_dict


def get_subj_lexicon():
    lexicon = load_file("lex.tff")
    subj_dict = build_subj_dicionary(lexicon)
    return subj_dict
```

```

def load_file(filename):
    file = open(filename, 'r')
    text = file.read()
    file.close()
    return text.split("\n")
```

```
def getlabelposneg(tweet):
  p = False
  n = False
  for word in tweet:
    x = word.lower().strip()
    if x in lex_dict:
      if lex_dict[x] == 'positive':
        p=True
      if lex_dict[x] == 'negative':
        n=True
  if p and n:
    return 1
  else:
    return 0
```
